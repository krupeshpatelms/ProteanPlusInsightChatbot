```python
import os

# Define directories
project_structure = [
    "backend/api", "backend/scripts",
    "frontend/components", "frontend/pages", "frontend/public",
    "data", "models", "vector_db", "config", "logs"
]

for directory in project_structure:
    os.makedirs(directory, exist_ok=True)

print("✔ Project structure created in Colab!")
```

```python
backend_path = "backend/api/main.py"

with open(backend_path, "w") as f:
    f.write('''
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def home():
    return {"message": "Protean AI Chatbot Backend Running"}

if __name__ == "__main__":
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
''')

print(f"✔ Backend API file created at {backend_path}")
```

```python
frontend_path = "frontend/App.js"

with open(frontend_path, "w") as f:
    f.write('''
import React from "react";

function App() {
    return (
        <div>
            <h1>Protean AI Chatbot Frontend Running</h1>
        </div>
    );
}

export default App;
''')

print(f"✔ Frontend App file created at {frontend_path}")
```
